```python
#missing data
df3 = pd.read_csv("nyc_weather.csv")
print(df3)
```

```python
print(df3.isnull().sum())
```

```python
#Replace() used to put values on NAN value
import numpy as np

#if we replace with one value then the whole NaN data value will replace which give inccorect answer, so for that we need to specify particular column data

df3["WindSpeedMPH"] = df3["WindSpeedMPH"].replace(np.nan, 6)
print(df3
     )
```

```python
df4 = df3
print(df4.isnull().sum())
```

```python
#Best way to replace the NaN values is find out mean of column and fill NaN value with mean value
#it wont effect the data too much
```

```python
#when we want to fill Nan values with backwar and forward value then
df5 = df4
df5["Events"] = df5["Events"].fillna(method = "bfill")  #backward fill
#df5["Events"] = df5["Events"].fillna(method = "bfill")  #forward fill
print(df5)
```

# Column Transformation

```python
#If we want to create new column in whic it shows details of "WindSpeedMPH" greate than 6, then loc() will gonna use

df6 = df5
df6.loc[(df6["WindSpeedMPH"] <= 6), "WindSpeed"] = "Normal Wind Speed"
df6.loc[(df6["WindSpeedMPH"] > 6), "WindSpeed"] = "High Wind Speed"
print(df6.head(10))
```

```python
df7 = df6
df7["Event + WindSpeed"] = df7["Events"].str.capitalize() +" "+ df7["WindSpeed"].str.capitalize()
print(df7)
```

```python
#if we want to fetch mathematical operations in new column then,

df8 = df7
df8["10% of Humidity"]  = (df8["Humidity"]/100)*10
print(df8)
```

# DataFrame COmparison

```python
dict = {"Fruits": ["mango","apple","banana","papaya"],
       "Price": [100,150,50,35],
       "Quantity": [15,10,10,3]}
df9 = pd.DataFrame(dict)
print(df9)
```

```python
df10 = df9.copy()
print(df10)
```

```python
#changing values
df10.loc[0,"Price"] = 120
df10.loc[1,"Price"] = 170
df10.loc[2,"Price"] = 150

df10.loc[0,"Quantity"] = 17
df10.loc[1,"Quantity"] = 12
df10.loc[2,"Quantity"] = 15

print(df10)
```

```python
print(df9.compare(df10))
```
